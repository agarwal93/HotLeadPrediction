```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
train_data=pd.read_excel('train_set.xlsx')
```

```python

pd.set_option('display.max_columns',None)
train_data.head()
```

```python
test_data=pd.read_excel('test_set.xlsx')
```

```python

pd.set_option('display.max_columns',None)
test_data.head()
```

```python
# taking all numerical features
```

```python
train_data1=train_data.drop(['Unnamed: 0','Asymmetrique Activity Score','Asymmetrique Profile Score'],axis=1)
```

```python
train_data1.head()
```

```python
test_data1=test_data.drop(['Unnamed: 0','Asymmetrique Activity Score','Asymmetrique Profile Score'],axis=1)
```

```python
test_data1.head()
```

```python
X_train=train_data1.drop('Converted',axis=1)
y_train=train_data1['Converted']
```

```python
X_test=test_data1.drop('Converted',axis=1)
y_test=test_data1['Converted']
```

```python
# now applying feature encoding
```

```python
X_train['train']=1
```

```python
X_test['train']=0
```

```python
X=pd.concat([X_train,X_test],axis=0)
```

```python
cat_cols=list(X.select_dtypes(include=['object']).columns)
```

```python
X=pd.get_dummies(X,columns=list(cat_cols),drop_first=True)
```

```python
X_train=X[X['train']==1]
```

```python
X_test=X[X['train']==0]
```

```python
X_train=X_train.drop('train',axis=1)
```

```python
X_test=X_test.drop('train',axis=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# applying RFECV for feature selection
```

```python
from sklearn.feature_selection import RFECV
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
from sklearn import metrics
```

```python
metrics.SCORERS.keys()
```

```python

rfecv=RFECV(model,cv=10,scoring='roc_auc')
rfecv.fit(X_train,y_train)
```

```python
X_train_selcols=X_train[list(X_train.columns[rfecv.support_])]
```

```python
X_test_selcols=X_test[list(X_train.columns[rfecv.support_])]
```

```python
len(X_train.columns)
```

```python
len(X_train_selcols.columns)
```

```python
len(X_test_selcols.columns)
```

```python
# feature scaling
```

```python
from sklearn.preprocessing import MinMaxScaler
mmc=MinMaxScaler()
```

```python
X_train_selcols=pd.DataFrame(mmc.fit_transform(X_train_selcols),columns=X_train_selcols.columns)
```

```python
X_test_selcols=pd.DataFrame(mmc.transform(X_test_selcols),columns=X_test_selcols.columns)
```

```python
X_train_selcols.head()
```

```python
X_test_selcols.head()
```

```python
# now applying basic Logistic Regression model
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train_selcols,y_train)
```

```python
y_test_pred=lr.predict(X_test_selcols)
y_test_predprob=lr.predict_proba(X_test_selcols)[:,1]
```

```python
from sklearn import metrics
```

```python
# printing accuracy score on train set after building basic logistc regression model

print('accuracy score =',lr.score(X_train_selcols,y_train))
```

```python
# printing all scores on test set after building basic logistc regression model

print('accuracy score =',lr.score(X_test_selcols,y_test))
print('roc_auc_score =',metrics.roc_auc_score(y_test,y_test_predprob))
print('precion score =',metrics.precision_score(y_test,y_test_pred))
print('recall score =',metrics.recall_score(y_test,y_test_pred))
print('f1 score =',metrics.f1_score(y_test,y_test_pred))
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds=roc_curve(y_test,y_test_predprob)
thresholds=thresholds[0]-1
plt.plot(fpr,tpr,'g')
plt.plot(fpr,fpr,'r')
plt.show()
```

```python
# now applying  random forest Classifier with hyperparameter tuning
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(random_state=10)
```

```python
from scipy.stats import randint as sp_randint
```

```python
params={'n_estimators':sp_randint(80,200),'max_depth':sp_randint(2,8),'min_samples_split':sp_randint(2,30),
       'min_samples_leaf':sp_randint(1,20)}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rsearch=RandomizedSearchCV(rfc,param_distributions=params,cv=10,scoring='roc_auc',n_iter=150,n_jobs=-1)
```

```python
rsearch.fit(X_train_selcols,y_train)
```

```python
rsearch.best_params_
```

```python
rfc1=RandomForestClassifier(random_state=10,**rsearch.best_params_)
```

```python
rfc1.fit(X_train_selcols,y_train)
```

```python
y_test_pred=rfc1.predict(X_test_selcols)
y_test_predprob=rfc1.predict_proba(X_test_selcols)[:,1]
```

```python
from sklearn import metrics
```

```python
# printing accuracy score on train set after building random forest classifier

print('accuracy score =',rfc1.score(X_train_selcols,y_train))
```

```python
# printing all scores on test set after building random forest classifier

print('accuracy score =',rfc1.score(X_test_selcols,y_test))
print('roc_auc_score =',metrics.roc_auc_score(y_test,y_test_predprob))
print('precion score =',metrics.precision_score(y_test,y_test_pred))
print('recall score =',metrics.recall_score(y_test,y_test_pred))
print('f1 score =',metrics.f1_score(y_test,y_test_pred))
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_test_predprob)
thresholds=thresholds[0]-1
plt.plot(fpr,tpr,'g')
plt.plot(fpr,fpr,'r')
plt.show()
```
